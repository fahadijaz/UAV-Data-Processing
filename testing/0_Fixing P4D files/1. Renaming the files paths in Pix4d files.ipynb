# LIST FILES IN FOLDER AND GET THEIR PATH AND IF FOUND THE FILE IN THE PIX4D FILE THEN UPDATE THE PATH


```python
# # Print all file names in a certain folder

# from os import walk

# # folder path
# dir_path = dir_path

# # list to store files name
# res = []
# for (dir_path, dir_names, file_names) in walk(dir_path):
#     res.extend(file_names)
# print(res)
```

```python
# List files with mentioned file type
import os
def list_files(filepath, filetype):
   paths = []
   for root, dirs, files in os.walk(filepath):
      for file in files:
         if file.lower().endswith(filetype.lower()):
            paths.append(os.path.join(root, file))
            # print(file)
   return(paths)
```

```python
# List files with all file types
import os
def list_files_all(filepath):
   paths = []
   for root, dirs, files in os.walk(filepath):
      for file in files:
        paths.append(os.path.join(root, file))
        # print(file)
   return(paths)
```

# Make a dict of files names and paths

```python
all_files = list_files_all(dir_path)
all_files
```

```python
delim = "\\"
dict_files = {}
for fil in all_files:
    corrected_path = fil.rsplit(delim, 1)[0].replace("\\","/") # modyifying the path to match the path syntax used in p4d files with /
    dict_files[fil.rsplit(delim, 1)[1]]=corrected_path

    # print(fil.rsplit(delim, 1)[1])
dict_files
```

# Use the files names to serch in the pix4d file and update their path

```python
# import re
# re.sub('\nThis.*?ok','',a, flags=re.DOTALL)
```

```python
# initializing substrings
files_in_p4d = []

sub1 = "<image path="
sub2 = " group="
 
s=str(re.escape(sub1))
 
e=str(re.escape(sub2))

res=re.findall(s+"(.*)"+e,filedata)
for ik in res:
    ik.strip("\"")
    files_in_p4d.append(ik.strip("\""))  # removing the " from the file path
print(len(res))
files_in_p4d
```

```python
delim1  = "/"

repl_dict = {}

for entry_p4d in files_in_p4d:
    filename_p4d = entry_p4d.rsplit(delim1, 1)[1]
    filepath_p4d =dict_files[filename_p4d] # separating the 

    repl_dict[entry_p4d] = filepath_p4d+"/"+filename_p4d
    
    # print(filepath_p4d+"/"+filename_p4d,"\n", entry_p4d)
    # print(filename_p4d[:-1])
```

# Replacing the file paths based on the repl_dict


```python
# Read in the file
with open(file_path, 'r') as file:
  filedata = file.read()
```

```python
# Replacing the file paths based on the repl_dict

for i in repl_dict:
    filedata = filedata.replace(i,repl_dict[i])
```

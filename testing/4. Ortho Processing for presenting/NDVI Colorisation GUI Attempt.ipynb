# Colorise the NDVI Orthomosaics

```python
import os
import shutil

# Define colormap categories
COLORMAP_CATEGORIES = {
    "Perceptually_Uniform (Best for Scientific Visualization)": ["viridis", "plasma", "inferno", "magma", "cividis"],
    "Classic Diverging (Good for Differences)": ["coolwarm", "RdYlBu", "RdYlGn", "PiYG", "Spectral", "bwr"],
    "Sequential Colormaps (For NDVI, Elevation, etc.)": ["Blues", "Greens", "Oranges", "Purples", "Reds"],
    "Rainbow & Jet Variants (Not Uniform, but Classic)": ["jet", "rainbow", "nipy_spectral"],
    "Cyclic (Best for Wind Direction, Phase Angles)": ["twilight", "twilight_shifted", "hsv"],
    "Categorical (For Class Labels)": ["tab10", "tab20", "Set1", "Set2", "Set3"],
    "Extra Colorcet Colormaps (for better color mapping)": ["fire", "glasbey", "rainbow4", "kb", "kr", "bgy", "bmw", "coolwarm_cet"]
}

def create_subfolders(base_folder, categories):
    """Create subfolders for each colormap category."""
    for category in categories:
        folder_path = os.path.join(base_folder, category)
        if not os.path.exists(folder_path):
            os.makedirs(folder_path)

def segregate_files(input_folder):
    """Segregate files into subfolders based on colormap name in the filename."""
    # Create subfolders for each category
    create_subfolders(input_folder, COLORMAP_CATEGORIES.keys())

    # Iterate through all files in the input folder
    for file_name in os.listdir(input_folder):
        # Check if the file is a .tif or _with_colorbar.png
        if file_name.endswith(".tif") or file_name.endswith("_with_colorbar.png"):
            # Find the category for this file
            for category, colormaps in COLORMAP_CATEGORIES.items():
                for colormap in colormaps:
                    if colormap in file_name:  # Check if colormap name is in the filename
                        # Move the file to the corresponding subfolder
                        src_path = os.path.join(input_folder, file_name)
                        dest_folder = os.path.join(input_folder, category)
                        dest_path = os.path.join(dest_folder, file_name)
                        shutil.move(src_path, dest_path)
                        print(f"Moved {file_name} to {category} folder.")
                        break
                else:
                    continue
                break
            else:
                print(f"‚ö†Ô∏è No category found for {file_name}.")

```

```python
# Define NDVI min and max for standardization
NDVI_MIN, NDVI_MAX = -1.0, 1.0  

# Define the colormap (e.g., 'viridis', 'jet', or custom)
COLORMAP = plt.cm.viridis

# Folder paths
input_folder = r"C:\Users\muij\OneDrive - Norwegian University of Life Sciences\Documents\ORTHO_2024\PRO_BAR_S√òR√ÖS\test"
output_folder = r"C:\Users\muij\OneDrive - Norwegian University of Life Sciences\Documents\ORTHO_2024\PRO_BAR_S√òR√ÖS\NDVI\test_color_small"
os.makedirs(output_folder, exist_ok=True)
```

```python
def normalize_ndvi(ndvi_array, vmin=NDVI_MIN, vmax=NDVI_MAX, nodata_value=None):
    """Normalize NDVI values to 0-1 scale, handling NoData values."""
    print(" - Normalizing NDVI values...")
    ndvi_array = ndvi_array.astype(np.float32)  # Ensure float calculations
    
    # Replace nodata values with NaN
    if nodata_value is not None:
        print(f"   - Handling NoData values ({nodata_value})...")
        ndvi_array[ndvi_array == nodata_value] = np.nan

    # Normalize (ignore NaNs)
    normalized = (ndvi_array - vmin) / (vmax - vmin)
    normalized = np.clip(normalized, 0, 1)  # Keep values in range

    return normalized

def apply_colormap(image_array, colormap=COLORMAP):
    """Apply colormap to normalized NDVI and return as RGB (0-255)."""
    print(" - Applying colormap...")
    normalized = normalize_ndvi(image_array)
    colored = colormap(normalized)[:, :, :3]  # Remove alpha channel
    return (colored * 255).astype(np.uint8)

def process_ndvi_tif(file_path, output_path):
    """Read NDVI TIFF, handle nodata, colorize it, and save as new TIFF."""
    print(f"\nProcessing: {os.path.basename(file_path)}")
    
    try:
        with rasterio.open(file_path) as src:
            print(" - Reading NDVI TIFF file...")
            ndvi = src.read(1).astype(np.float32)  # Read as float to handle NaNs
            nodata_value = src.nodata  # Get NoData value from metadata
            print(f"   - NoData value detected: {nodata_value}")

            # Apply colormap
            rgb_array = apply_colormap(ndvi)

            # Update metadata for RGB output
            profile = src.profile
            profile.update(
                dtype=rasterio.uint8,
                count=3,  # Three bands for RGB
                photometric="RGB",
                nodata=None  # Remove NoData for RGB
            )

            print(" - Saving colorized TIFF...")
            with rasterio.open(output_path, "w", **profile) as dst:
                for i in range(3):
                    dst.write(rgb_array[:, :, i], i + 1)

            print(f"‚úÖ Successfully processed: {os.path.basename(file_path)}")

    except Exception as e:
        print(f"‚ùå ERROR processing {file_path}: {str(e)}")
```

# Downscale the Orthomosaics

```python
import os
import rasterio
from rasterio.enums import Resampling
import numpy as np
import matplotlib.pyplot as plt
```

```python
def downscale_tif(file_path, output_path, scale_factor):
    """Downscale an orthomosaic TIFF by a given scale factor."""
    print(f"\nProcessing: {os.path.basename(file_path)} at {int(scale_factor * 100)}% resolution")

    try:
        with rasterio.open(file_path) as src:
            # Compute new dimensions
            new_width = int(src.width * scale_factor)
            new_height = int(src.height * scale_factor)
            print(f" - Original size: {src.width} x {src.height}")
            print(f" - New size: {new_width} x {new_height}")

            # Read and resample image
            resampled_array = src.read(
                out_shape=(src.count, new_height, new_width),
                resampling=Resampling.bilinear  # Bilinear for smooth downscaling
            )

            # Update metadata
            profile = src.profile
            profile.update(
                width=new_width,
                height=new_height,
                transform=src.transform * src.transform.scale(
                    (src.width / new_width), (src.height / new_height)
                )
            )

            # Create output filename
            scale_label = f"{int(scale_factor * 100)}%"
            output_scaled_path = output_path.replace(".tif", f"_{scale_label}.tif")

            # Save downscaled image
            print(f" - Saving downscaled TIFF: {output_scaled_path}")
            with rasterio.open(output_scaled_path, "w", **profile) as dst:
                dst.write(resampled_array)

            print(f"‚úÖ Successfully downscaled: {os.path.basename(file_path)} to {scale_label}")

    except Exception as e:
        print(f"‚ùå ERROR processing {file_path}: {str(e)}")
```

```python
import os
import rasterio
from rasterio.enums import Resampling
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image  # For JPG conversion

# Folder paths (UPDATE THESE)
input_folder = "path/to/ndvi_tifs"
output_folder = "path/to/output_downscaled"
os.makedirs(output_folder, exist_ok=True)

# Predefined scaling options
SCALE_OPTIONS = {
    "50%": 0.5,
    "25%": 0.25,
    "10%": 0.1,
}

def downscale_tif(file_path, output_path, scale_factor, convert_to_jpg=False):
    """Downscale an orthomosaic TIFF by a given scale factor and optionally convert to JPG."""
    print(f"\nProcessing: {os.path.basename(file_path)} at {int(scale_factor * 100)}% resolution")

    try:
        with rasterio.open(file_path) as src:
            # Compute new dimensions
            new_width = int(src.width * scale_factor)
            new_height = int(src.height * scale_factor)
            print(f" - Original size: {src.width} x {src.height}")
            print(f" - New size: {new_width} x {new_height}")

            # Read and resample image
            resampled_array = src.read(
                out_shape=(src.count, new_height, new_width),
                resampling=Resampling.bilinear  # Bilinear for smooth downscaling
            )

            # Update metadata
            profile = src.profile
            profile.update(
                width=new_width,
                height=new_height,
                transform=src.transform * src.transform.scale(
                    (src.width / new_width), (src.height / new_height)
                )
            )

            # Create output filename
            scale_label = f"{int(scale_factor * 100)}%"
            output_scaled_path = output_path.replace(".tif", f"_{scale_label}.tif")

            # Save downscaled image as TIFF
            print(f" - Saving downscaled TIFF: {output_scaled_path}")
            with rasterio.open(output_scaled_path, "w", **profile) as dst:
                dst.write(resampled_array)

            # Convert to JPG if requested
            if convert_to_jpg:
                jpg_path = output_scaled_path.replace(".tif", ".jpg")
                convert_tif_to_jpg(output_scaled_path, jpg_path)

            print(f"‚úÖ Successfully downscaled: {os.path.basename(file_path)} to {scale_label}")

    except Exception as e:
        print(f"‚ùå ERROR processing {file_path}: {str(e)}")

def convert_tif_to_jpg(tif_path, jpg_path):
    """Convert a TIFF file to JPG format."""
    try:
        print(f" - Converting to JPG: {jpg_path}")
        with rasterio.open(tif_path) as src:
            array = src.read(1)  # Read first band for grayscale

            # Normalize for visualization
            array = (array - np.min(array)) / (np.max(array) - np.min(array))
            array = (array * 255).astype(np.uint8)  # Convert to 8-bit grayscale

            # Save as JPG
            img = Image.fromarray(array)
            img.save(jpg_path, "JPEG", quality=95)
            print(f"‚úÖ JPG saved: {jpg_path}")

    except Exception as e:
        print(f"‚ùå ERROR converting {tif_path} to JPG: {str(e)}")

def preview_downscaling(file_path):
    """Preview different downscaling options using the first image."""
    print("\nüîç Previewing downscaling options...")
    
    with rasterio.open(file_path) as src:
        fig, axes = plt.subplots(1, len(SCALE_OPTIONS), figsize=(15, 5))

        for ax, (label, scale) in zip(axes, SCALE_OPTIONS.items()):
            # Downscale using the selected factor
            preview_array = src.read(
                out_shape=(1, int(src.height * scale), int(src.width * scale)),
                resampling=Resampling.bilinear
            )[0]  # Extract first band for visualization
            
            # Normalize for visualization
            preview_array = (preview_array - np.min(preview_array)) / (np.max(preview_array) - np.min(preview_array))
            
            ax.imshow(preview_array, cmap="gray")
            ax.set_title(label)
            ax.axis("off")

        plt.show()

# Step 1: Preview the first image at different resolutions
tif_files = [f for f in os.listdir(input_folder) if f.endswith(".tif")]
if not tif_files:
    print("‚ùå No TIFF files found in the input folder.")
    exit()

first_image = os.path.join(input_folder, tif_files[0])
preview_downscaling(first_image)

# Step 2: Ask the user for resolution choice
print("\nAvailable downscaling options:", ", ".join(SCALE_OPTIONS.keys()))
selected_scales = input("Enter scale percentages (comma-separated) or press Enter for 50% default: ").strip()

if not selected_scales:
    selected_scales = ["50%"]
else:
    selected_scales = [s.strip() for s in selected_scales.split(",") if s.strip() in SCALE_OPTIONS]

selected_factors = [SCALE_OPTIONS[s] for s in selected_scales]

# Step 3: Ask the user if they want JPG conversion
convert_to_jpg = input("Do you want to convert results to JPG? (yes/no): ").strip().lower() == "yes"

print(f"\nüìè Selected downscaling: {', '.join(selected_scales)}")
print(f"üì∏ Convert to JPG: {'Yes' if convert_to_jpg else 'No'}")

# Step 4: Process all TIFF images with the selected scaling factors
print("\nüöÄ Starting Downscaling Process...\n")
for filename in tif_files:
    input_path = os.path.join(input_folder, filename)
    for scale in selected_factors:
        output_path = os.path.join(output_folder, filename)
        downscale_tif(input_path, output_path, scale, convert_to_jpg)

print("\n‚úÖ Downscaling complete for all images!\n")
```

```python


def preview_colormaps(file_path, colormaps, rows=4, cols=4):
    """Display an image with multiple colormaps for user selection."""
    print(f"\nüîç Previewing: {os.path.basename(file_path)} with different colormaps...")

    with rasterio.open(file_path) as src:
        band1 = src.read(1)
        band1 = (band1 - np.min(band1)) / (np.max(band1) - np.min(band1))  # Normalize

    num_colormaps = len(colormaps)
    fig, axes = plt.subplots(rows, cols, figsize=(cols * 4, rows * 3))
    axes = axes.ravel()

    for i, (cmap_name, cmap) in enumerate(colormaps.items()):
        if i >= len(axes): break
        axes[i].imshow(band1, cmap=cmap)
        axes[i].set_title(cmap_name, fontsize=10)
        axes[i].axis("off")

    for j in range(i + 1, len(axes)): fig.delaxes(axes[j])  # Remove extra subplots
    plt.tight_layout()
    plt.show()

def downscale_tif(file_path, output_path, scale_factor, convert_to_jpg):
    """Downscale an orthomosaic and optionally convert to JPG."""
    print(f"üìè Resizing {os.path.basename(file_path)} to {int(scale_factor * 100)}%")
    
    try:
        with rasterio.open(file_path) as src:
            new_width, new_height = int(src.width * scale_factor), int(src.height * scale_factor)
            resampled_array = src.read(out_shape=(src.count, new_height, new_width), resampling=Resampling.bilinear)

            profile = src.profile
            profile.update(width=new_width, height=new_height, transform=src.transform * src.transform.scale(
                (src.width / new_width), (src.height / new_height)))

            output_scaled_path = output_path.replace(".tif", f"_{int(scale_factor * 100)}%.tif")
            print(f" - Saving {output_scaled_path}")
            with rasterio.open(output_scaled_path, "w", **profile) as dst:
                dst.write(resampled_array)

            if convert_to_jpg:
                jpg_path = output_scaled_path.replace(".tif", ".jpg")
                convert_tif_to_jpg(output_scaled_path, jpg_path)
    except Exception as e:
        print(f"‚ùå ERROR: {str(e)}")

def convert_tif_to_jpg(tif_path, jpg_path):
    """Convert TIFF to JPG."""
    try:
        print(f" - Converting {os.path.basename(tif_path)} to JPG")
        with rasterio.open(tif_path) as src:
            array = src.read(1)
            array = (array - np.min(array)) / (np.max(array) - np.min(array)) * 255
            Image.fromarray(array.astype(np.uint8)).save(jpg_path, "JPEG", quality=95)
            print(f"‚úÖ JPG saved: {jpg_path}")
    except Exception as e:
        print(f"‚ùå ERROR converting {tif_path} to JPG: {str(e)}")

```

```python
# Function to find the tif files in a given folderÂ¶
def find_files_in_folder(folder_path, extension=None, recursive=False):
    """
    Retrieves a list of file paths in a specified folder, optionally filtered by file extension,
    and optionally including subdirectories.

    Parameters:
    - folder_path (str): The path of the folder to search for files.
    - extension (str, optional): The file extension to filter by (e.g., "txt" or "tif").
                                 If None, the function lists all files.
    - recursive (bool, optional): If True, includes files from all subdirectories within `folder_path`.
                                  Defaults to False (only lists files in the specified folder).

    Returns:
    - list of str: A list of file paths that match the specified extension in the folder.
                   If no matching files are found, returns a list containing an empty string.

    Example:
    >>> find_files_in_folder("/path/to/folder", "tif")
    ['/path/to/folder/file1.tif', '/path/to/folder/file2.tif']

    >>> find_files_in_folder("/path/to/folder", recursive=True)
    ['/path/to/folder/file1.tif', '/path/to/folder/subfolder/file2.txt', '/path/to/folder/file3.jpg']
    """
    
    matched_files = []

    # Determine the search pattern based on whether an extension is provided and recursion is enabled
    if extension:
        if recursive:
            search_pattern = os.path.join(folder_path, f"**/*.{extension}")
        else:
            search_pattern = os.path.join(folder_path, f"*.{extension}")
    else:
        # No extension specified, handle both recursive and non-recursive cases
        if recursive:
            search_pattern = os.path.join(folder_path, "**/*")
        else:
            search_pattern = os.path.join(folder_path, "*")
    
    # Use glob to find matching files in the specified directory and subdirectories if recursive
    matched_files.extend(glob.glob(search_pattern, recursive=recursive))
    
    # If no files are found, return a list with an empty string
    if not matched_files:
        matched_files = [""]

    return matched_files
```

```python
import geopandas as gpd
import rasterio
import numpy as np
import pandas as pd
from rasterio.mask import mask

def calculate_zonal_statistics(geojson_file, orthomosaic_file, output_csv):
    """
    Calculate zonal statistics (mean, median, stddev) for each polygon in the GeoJSON file
    using the corresponding values from the orthomosaic raster.

    Parameters:
        geojson_file (str): Path to the GeoJSON file containing the polygons.
        orthomosaic_file (str): Path to the orthomosaic raster file.
        output_csv (str): Path to the output CSV file to save the results.

    Returns:
        None
    """
    # Step 1: Load the GeoJSON polygons
    # gdf = gpd.read_file(geojson_file)
    gdf = geojson_file

    # Step 2: Load the orthomosaic (raster) using rasterio
    with rasterio.open(orthomosaic_file) as src:
        # Prepare the list to store the statistics for each polygon
        stats = []

        # Step 3: Loop through each polygon in the GeoJSON file
        for _, row in gdf.iterrows():
            polygon = row['geometry']

            # Step 4: Mask the orthomosaic using the polygon geometry
            # The `mask` function extracts the raster data corresponding to the polygon
            try:
                out_image, out_transform = mask(src, [polygon], crop=True)
                out_image = out_image[0]  # Take the first band (assuming single band raster)

                # Masking may leave NaN or masked values, so we need to handle that
                valid_pixels = out_image[~np.isnan(out_image)]

                if valid_pixels.size > 0:  # Check if there are valid pixels
                    mean = np.mean(valid_pixels)
                    median = np.median(valid_pixels)
                    stddev = np.std(valid_pixels)

                    # Append the statistics for this polygon
                    stats.append({
                        'ID': row['ID'],
                        'Mean': mean,
                        'Median': median,
                        'StdDev': stddev
                    })
                else:
                    # In case there are no valid pixels (e.g., no overlap with raster)
                    stats.append({
                        'ID': row['ID'],
                        'Mean': None,
                        'Median': None,
                        'StdDev': None
                    })

            except Exception as e:
                print(f"Error processing polygon {row['ID']}: {e}")
                stats.append({
                    'ID': row['ID'],
                    'Mean': None,
                    'Median': None,
                    'StdDev': None
                })

        # Step 5: Create a DataFrame from the results
        stats_df = pd.DataFrame(stats)

        # Step 6: Save the results to a CSV
        stats_df.to_csv(output_csv, index=False)
        print(f"Zonal statistics saved to {output_csv}")


```

```python
import geopandas as gpd
import rasterio

def check_and_reproject(gdf, raster_path):
    """
    Check the CRS of the GeoDataFrame (gdf) and reproject if necessary to match the CRS of the raster.
    
    Parameters:
        gdf (GeoDataFrame): The GeoDataFrame of polygons.
        raster_path (str): Path to the raster file.
        
    Returns:
        GeoDataFrame: Reprojected GeoDataFrame if CRS was different, else returns the original.
    """
    # Open the raster file to get its CRS
    with rasterio.open(raster_path) as src:
        raster_crs = src.crs
    
    # Check if the CRS of the GeoDataFrame matches the raster's CRS
    if gdf.crs != raster_crs:
        print(f"CRS mismatch: Reprojecting GeoDataFrame from {gdf.crs} to {raster_crs}.")
        gdf = gdf.to_crs(raster_crs)  # Reproject GeoDataFrame to match raster CRS
    else:
        print("CRS match: No reprojecting needed.")
    
    return gdf

```

```python
import geopandas as gpd
import rasterio
import numpy as np
import pandas as pd
import os

# Set working directory
os.chdir("P:/PhenoCrop")

# Load Polygon GeoJSON file
shp = gpd.read_file("P:/PhenoCrop/3_qgis/shape files new/Phenocrop.geojson")

# Plot the polygons (optional, requires matplotlib)
# shp.plot()

# Spectral data upload
r1_date = "P:/PhenoCrop/Phenotyping 2024/PHENO/MS/20240624 PHENO P4M 20m MS/band_data"
rlist = [os.path.join(r1_date, f) for f in os.listdir(r1_date) if "index" in f]

# Stack image files (load rasters)
rasters = []
for raster_file in rlist:
    rasters.append(rasterio.open(raster_file))

# Assign names to raster bands
band_names = ['blue', 'green', 'ndvi', 'nir', 'rededge', 'red']
for idx, raster in enumerate(rasters):
    raster.name = band_names[idx]

# Function to extract statistics from raster based on polygons
def extract_statistics(rasters, polygons, bands):
    # Prepare lists to store results
    results = {band: {'median': [], 'mean': [], 'std': []} for band in bands}

    # Loop through rasters
    for raster, band in zip(rasters, bands):
        # Read raster data
        band_data = raster.read(1)  # Assuming single-band rasters

        # Loop through polygons (geometries)
        for geometry in polygons.geometry:
            # Mask raster with polygon (getting the pixels inside the polygon)
            mask, transform = rasterio.mask.mask(raster, [geometry], crop=True)
            
            # Flatten the array and remove the 'no data' values (assuming a value of -9999)
            mask = mask[mask != raster.nodata]
            
            # Calculate statistics
            results[band]['median'].append(np.median(mask))
            results[band]['mean'].append(np.mean(mask))
            results[band]['std'].append(np.std(mask))

    return results

# Extract statistics for each band
results = extract_statistics(rasters, shp, band_names)

# Convert results to DataFrame
df = pd.DataFrame(results)

# Prepare the final dataframe for each polygon
VI = pd.DataFrame()

# Add the date column (assuming that each date corresponds to a raster)
Flight = [os.path.basename(r).split('_')[-3] for r in rlist]
Flight = [date[-15:] for date in Flight]  # Extract last 15 characters (date)

# Loop to create column names and add values
for band in band_names:
    for stat in ['median', 'mean', 'std']:
        VI[f'{band}_{stat}'] = df[band][stat]

# Add Flight Date to the DataFrame
VI['Flight_Date'] = Flight

# Save the results to CSV
VI.to_csv("P:/PhenoCrop/Phenotyping 2024/PHENO/MS/20240624 PHENO P4M 20m MS/R_generated_values.csv", index=False)

print("Process completed and saved to CSV.")
```

```python
import geopandas as gpd
import rasterio
import numpy as np
import pandas as pd
from rasterio.mask import mask
from rasterio import plot

# Load Polygon GeoJSON file
shp = gpd.read_file(geojson_file)
shp = gdf
shp.plot(edgecolor='red')
# Plot the polygons (optional, requires matplotlib)
# shp.plot()

# Spectral data upload
rlist = new_path

# Load the raster stack (multiband raster stack)
rasters = [rasterio.open(f) for f in rlist]

# Assign band names manually
band_names = ['blue', 'green', 'nir', 'red_edge', 'red']

# Create empty DataFrames to store results
VI = {
    # 'NDVI_med': [], 'NDVI_mean': [], 'NDVI_std': [],
    'blue_med': [], 'blue_mean': [], 'blue_std': [],
    'red_med': [], 'red_mean': [], 'red_std': [],
    'green_med': [], 'green_mean': [], 'green_std': [],
    'redE_med': [], 'redE_mean': [], 'redE_std': [],
    'nir_med': [], 'nir_mean': [], 'nir_std': []
}

# Flight dates (assuming these are available in the filenames or specified manually)
Flight = [os.path.basename(f)[-15:] for f in rlist]  # Extract last 15 characters (example)

# Function to calculate the Zonal Statistics (mean, median, std) from the raster and shapefile
import numpy as np
import rasterio
from rasterio.mask import mask

def extract_zonal_stats(raster, shapes, band_index):
    """
    Extract mean, median, and standard deviation values from the raster for the polygons in the shapefile.
    Args:
        raster (rasterio dataset): Raster dataset (e.g. blue, green, etc.)
        shapes (GeoDataFrame): Shapefile containing the polygons
        band_index (int): Index for the raster band (0-indexed)
    Returns:
        tuple: (mean, median, std) for the given band and shapefile
    """
    # Check the number of bands in the raster
    num_bands = raster.count
    
    # Handle single-band raster by forcing band_index = 0
    if num_bands == 1:
        band_index = 0
    
    # Ensure the band_index is valid
    if band_index >= num_bands:
        raise IndexError(f"Band index {band_index} is out of bounds for raster with {num_bands} bands.")
    
    # Mask raster by the shapefile geometry
    out_image, out_transform = mask(raster, shapes.geometry, crop=True)
    
    # Extract the specific band from the multi-band raster (usually the first band is at index 0)
    out_image_band = out_image[band_index, :, :]  # Select the specific band (adjust index)
    
    # Get the NoData value for the raster
    nodata_value = raster.nodata
    
    # Mask out the NoData values
    if nodata_value is not None:
        out_image_band = np.ma.masked_equal(out_image_band, nodata_value)
    
    # Compute statistics, excluding masked (NoData) values
    pixel_values = out_image_band.flatten()
    
    if len(pixel_values) == 0 or np.ma.count_masked(out_image_band) == len(pixel_values):
        return np.nan, np.nan, np.nan
    
    return np.mean(pixel_values), np.median(pixel_values), np.std(pixel_values)


# Extract spectral values and store in the DataFrame
for band_index, band_name in enumerate(band_names):
    for raster in rasters:
        # Extract the statistics (mean, median, std) for each polygon
        print(raster)
        mean, median, std = extract_zonal_stats(raster, shp, band_index)
        
        # Store the results in the DataFrame
        VI[f'{band_name}_mean'].append(mean)
        VI[f'{band_name}_med'].append(median)
        VI[f'{band_name}_std'].append(std)

# Create the DataFrame from the results
VI_df = pd.DataFrame(VI)

# Assign column names with the flight dates (e.g., '20240624_PHENO_P4M_20m_MS_NDVImean')
for band in band_names:
    VI_df.columns = [f'{flight}_{band}_mean' for flight in Flight]
    VI_df.columns = [f'{flight}_{band}_med' for flight in Flight]
    VI_df.columns = [f'{flight}_{band}_std' for flight in Flight]

# Save the DataFrame to a CSV file
output_file = "_R_generated_values.csv"
VI_df.to_csv(output_file, index=False)

print(f"Results saved to {output_file}")
```
